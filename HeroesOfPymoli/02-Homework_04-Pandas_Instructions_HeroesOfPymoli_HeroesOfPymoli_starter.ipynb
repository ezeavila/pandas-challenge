```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/02-Homework_04-Pandas_Instructions_HeroesOfPymoli_Resources_purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
general_df = pd.read_csv(file_to_load)

general_df.head()
```

## Player Count

* Display the total number of player

```python
player_df = general_df.loc[: , ["Gender", "SN", "Age"]]

player_df = player_df.drop_duplicates()

numberplayers= player_df.count()[0] #We use zero to only shoe the first value of the df

pd.DataFrame({"Total of Players" : [numberplayers]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Total_sales = general_df["Price"].sum()
Average_Price = general_df["Price"].mean()
Purchase_Count = general_df["Price"].count()
Item_Count = len(general_df["Item ID"].unique()) #if I don't include the len function, it won't give me the total count of elements

summary_table = pd.DataFrame({"Number of Unique Items" : Item_Count ,
                             "Total Revenue" : [Total_sales] ,
                             "Number of Purchase" : [Purchase_Count] ,
                             "Average Price" : [Average_Price] })

summary_table["Average Price"] = summary_table["Average Price"].map("${:,.2f}".format) #apply on rest of the columns

summary_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Value count on column "Gender" and create a new column divifind the value counts by the total number of observations
```

```python
gender_df = player_df["Gender"].value_counts()

gender_perct = gender_df / numberplayers

gender_perct

gender_demo = pd.DataFrame({"Total Count" : gender_df , "Percentage of Players" : gender_perct})

gender_demo

#apply the map function on the percentajes .2 and % percentage
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count = general_df.groupby("Gender").count()["Price"]
gender_average = general_df.groupby("Gender").mean()["Price"]
gender_totalpurchase = general_df.groupby("Gender").sum()["Price"]
average_perperson = gender_totalpurchase / gender_demo["Total Count"]

gender_summary = pd.DataFrame({"Purchase Count" : purchase_count,
                               "Average Purchase Price" : gender_average , 
                              "Total Purchase Price" : gender_totalpurchase ,
                              "Avg Total Purchase per Person" : average_perperson})

gender_summary

#Will format later (we need to do one at a time)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0 , 9 , 14 , 19 , 24, 29 , 34 , 39 , 200]
age_groups = ["<10" , "10 to 14" , "15 to 19" , "20 to 24" , "25 to 29" , "30 to 34" , "35 to 39" , "40+"]

player_df["Age Range"] = pd.cut(player_df["Age"], bins,labels=age_groups)

player_df.groupby("Age Range").size()
```

```python
agerange_df = player_df["Age Range"].value_counts()
percentage_agerange =  agerange_df / numberplayers

age_range_summary = pd.DataFrame({"Total Count by Age" : agerange_df,
                                  "Percentage of Players" : percentage_agerange})


age_range_summary

#format cell , including the sorting of the rows
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
general_df["Age Range"] = pd.cut(general_df["Age"], bins,labels=age_groups)

purchase_count_by_age = general_df.groupby("Age Range").count()["Price"]
age_group_average = general_df.groupby("Age Range").mean()["Price"]
age_group_totalpurchase = general_df.groupby("Age Range").sum()["Price"]
average_peragegroup = age_group_totalpurchase / age_range_summary["Total Count by Age"]


age_range_summary_purchases = pd.DataFrame({"Number Purchases by Age Group" : purchase_count_by_age , 
                                           "Avg Purchase Price" : age_group_average,
                                           "Total Revenue by Age Group" : age_group_totalpurchase ,
                                           "Avg Revenue per Person" : average_peragegroup})

age_range_summary_purchases
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_by_SN = general_df.groupby("SN").count()["Purchase ID"]
avg_purchase_price = general_df.groupby("SN").mean() ["Price"]
total_purchase_value = purchase_by_SN * avg_purchase_price


Purchases_by_SN = pd.DataFrame({"Number Purchase Count" : purchase_by_SN , 
                                           "Avg Purchase Price by SN" : avg_purchase_price,
                                           "Total Purchase Value by SN" : total_purchase_value})

Purchases_by_SN = Purchases_by_SN.sort_values((['Number Purchase Count']), ascending = "false")

#sort from highest to lowest

Purchases_by_SN
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_by_title = general_df.groupby("Item Name").count()["Purchase ID"]
price_title = general_df.groupby("Item Name").mean() ["Price"]
total_purchase_title = purchase_by_title * price_title


purchases_by_title_df = pd.DataFrame({"Purchase Count Title" : purchase_by_title , 
                                           "Title Price" : price_title,
                                           "Total Revenue by Title" : total_purchase_title})

purchases_by_title_df = purchases_by_title_df.sort_values((['Purchase Count Title']), ascending = "false")

#sort from highest to lowest

purchases_by_title_df
```
